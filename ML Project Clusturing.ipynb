# Problem 2:
How many Unique patterns that exist in the historical stock data set, based on
fluctuations in price.

```python
from numpy import vstack,array
from numpy.random import rand
import numpy as np
from scipy.cluster.vq import kmeans,vq
import pandas as pd
from math import sqrt
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

```python
returns = df.pct_change().mean()*252
returns = pd.DataFrame(returns)
```

```python
returns
```

```python
returns.columns = ['Returns']
returns.Volatility = df.pct_change().std()*sqrt(252)
```

```python
# Making data into numpy array to feed into K-means.
df_2 = np.asarray([np.asarray(returns.Returns), np.asarray(returns.Volatility)]).T

X = df_2
distortions =[]
for k in range(2,20):
    k_means = KMeans(n_clusters = k)
    k_means.fit(X)
    distortions.append(k_means.inertia_)
```

```python
fig = plt.figure(figsize=(15,5))
plt.plot(range(2,20),distortions)
plt.grid(True)
plt.title('Elbow-Curve')
plt.show()
```

```python
from pylab import plot, show
#Lets compute kmeans with K =10 (10 clusters)
centroids,_= kmeans(df_2, 10)
```

```python
idx,_ = vq(df_2, centroids)
```

```python
plot(df_2[idx==0,0],df_2[idx==0,1],'ob',
     df_2[idx==1,0],df_2[idx==1,1],'oy',
     df_2[idx==2,0],df_2[idx==2,1],'or',
     df_2[idx==3,0],df_2[idx==3,1],'og',
     df_2[idx==4,0],df_2[idx==4,1],'om')

plot(centroids[:,0],centroids[:,1],'sg',markersize=10)
show()
```

```python
details = [(name,cluster) for name, cluster in zip(returns.index,idx)]
details[:5]
```

```python
df = pd.DataFrame(details, columns=['Stock_Name','Cluster_No'])
df.Cluster_No.value_counts()
```

```python
import seaborn as sns
ind = df.Cluster_No.value_counts()
plt.figure(figsize=(12,4))
sns.barplot(x=ind.index, y=ind.values, alpha=0.6)
plt.ylabel('Number of Occurences',fontsize=12)
plt.xlabel('Cluster Number',fontsize=12)
plt.title('Barplot - Indicator',fontsize =16)
plt.show()
```

# Problem 3:
Identify which all stocks are moving together and which all stocks are different from each other.

```python
df1
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# converting data into numpy arrays
X = df1.values

# Scaling the values
sc = StandardScaler()
X_scaled = sc.fit_transform(X)
```

```python
X_scaled
```

```python
pca = PCA(n_components=150)
pca.fit(X_scaled)

# the amount of variance that each PCA explains
var = pca.explained_variance_ratio_
```

```python
# cumulative varince explains
var1 = np.cumsum(np.round(var, decimals=4)*100)
print(var1)
```

```python
plt.plot(var1)
```

```python
comp = pd.DataFrame(pca.components_, columns = df1.columns)
comp.head()
```
